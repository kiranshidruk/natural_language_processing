```python
# remove HTML tags present in text

import re

def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

```python
#  removing URL's present in the text

def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')

    return pattern.sub(r'', text)
```

```python
#  remove punctuations in data

import string, time

exclude = string.punctuation
```

```python
def remove_punctuation(text):
    for char in exclude:
      text  = text.replace(char,'')
    return text
```

```python
test = "Abcd. how are you ? doing good!"
```

```python
print(remove_punctuation(test))
```

```python
df['review'] = df['review'].apply(remove_punctuation)
```

```python
#  this above technique is slow, lets try to do it in optimize way

def remove_punctuation1(text):
  return text.translate(str.maketrans('','',exclude))
```

```python
df['review'] = df['review'].apply(remove_punctuation)
```

```python
#  Chat word treatment ---> replacing short form word like FYI, ASAP with their full time.

import json

# Path to your JSON file
file_path = "chat-acronyms.json"

# Open and load JSON file
with open(file_path, "r") as file:
    data = json.load(file)

# 'data' is now a Python dictionary
print(type(data))
print(data)
```

```python
merged_dict = {item['Abbreviation']: item['Definition'] for item in data}

print(merged_dict)
```

```python
def char_conversion(text):
  new_text = []

  for word in text.split():
    if word.upper() in merged_dict:
      new_text.append(merged_dict[word.upper()])
    else:
      new_text.append(word)

  return " ".join(new_text)
```

```python
test = "hey are will be BRB"
```

```python
print(char_conversion(test))
```

```python
df['review'] = df['review'].apply(char_conversion)
```

```python
#  Removing the stopwords

from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)

    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['review'].apply(remove_stopwords)
```

```python
#  removing imogies, either we can replace it with a word or remove it

#  remove it
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
remove_text = "abc xyc usf "
```

```python
remove_emoji(remove_text)
```

```python
df['review'].apply(remove_emoji)
```

```python
#  Stemming

from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()

def stem(text):
  new_text = []
  for word in text.split():
    new_text.append(ps.stem(word))

  return " ".join(new_text)
```

```python
df['review'] = df['review'].apply(stem)
```

```python
#  lemmitization

#  it goes and checks in the dictionary



from nltk.stem import WordNetLemmatizer

wordnet_lemmatizer = WordNetLemmatizer()

sentence = "Hi, I'm Kiran Shidruk, a Machine Learning Engineer with over 5 years? of experience building production ML systems on cloud platforms like GCP and AWS. I specialize in taking models from research all the way through to deployment!, with expertise in threat detection, NLP, and computer vision!"
punctuations = "?:!.,;"
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word", "Lemma"))
for word in sentence_words:
    print("{0:20}{1:20}".format(word, wordnet_lemmatizer.lemmatize(word, pos='v')))
```
