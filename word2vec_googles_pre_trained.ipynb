```python
import kagglehub

path = kagglehub.dataset_download("leadbest/googlenewsvectorsnegative300")

print("Path to dataset files:", path)
```

```python
model = KeyedVectors.load_word2vec_format('/root/.cache/kagglehub/datasets/leadbest/googlenewsvectorsnegative300/versions/2/GoogleNews-vectors-negative300.bin.gz',binary=True,limit=500000)
```
