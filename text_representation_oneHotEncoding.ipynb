```python
import re

def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

```python
df['review'].apply(remove_html_tags)
```

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')

    return pattern.sub(r'', text)
```

```python
df['review'].apply(remove_url)
```

```python
#  remove punctuations in data

import string, time

exclude = string.punctuation
```

```python
def remove_punctuation(text):
    for char in exclude:
      text  = text.replace(char,'')
    return text
```

```python
df['review'].apply(remove_punctuation)
```

```python
#  this above technique is slow, lets try to do it in optimize way

def remove_punctuation1(text):
  return text.translate(str.maketrans('','',exclude))
```

```python
df['review'] = df['review'].apply(remove_punctuation)
```

```python

import json

# Path to your JSON file
file_path = "chat-acronyms.json"

# Open and load JSON file
with open(file_path, "r") as file:
    data = json.load(file)

# 'data' is now a Python dictionary
print(type(data))
print(data)
```

```python
merged_dict = {item['Abbreviation']: item['Definition'] for item in data}

print(merged_dict)
```

```python
def char_conversion(text):
  new_text = []

  for word in text.split():
    if word.upper() in merged_dict:
      new_text.append(merged_dict[word.upper()])
    else:
      new_text.append(word)

  return " ".join(new_text)
```

```python
df['review'].apply(char_conversion)
```

```python
#  Removing the stopwords

from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)

    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
df['review'].apply(remove_stopwords)
```

```python
#  removing imogies, either we can replace it with a word or remove it

#  remove it
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
df['review'].apply(remove_emoji)
```

```python
#  Stemming

from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()

def stem(text):
  new_text = []
  for word in text.split():
    new_text.append(ps.stem(word))

  return " ".join(new_text)
```

```python
df['review'].apply(stem)
```

```python
#  Find out the number of words in the entire corpus and also the total number of unique words(vocabulary) using just python

corpus = ' '.join(df['review'])

words = corpus.split()

words = [w.lower() for w in words]

total_words = len(words)

unique_words = len(set(words))

print("Total number of words in corpus:", total_words)
print("Total number of unique words (vocabulary):", unique_words)
```

```python
import re

# Build vocabulary from it
vocab = set()
for review in df['review']:
    words = review.split()
    vocab.update(words)

# Now vocab is a set of all unique words (the vocabulary)
vocab = sorted(list(vocab))
```

```python
def one_hot_encode(review, vocab):
    words = review.split()
    return [1 if word in words else 0 for word in vocab]

encoded_reviews = df['review'].apply(lambda x: one_hot_encode(x, vocab))
encoded_df = pd.DataFrame(encoded_reviews.tolist(), columns=vocab)
print(encoded_df.head())
```
